```python
bank_data = pd.read_csv("./bank_v1.csv")
```

```python
column_list = bank_data.columns
```

```python
column_list = [u'age', u'balance', u'day', u'duration', u'campaign', u'pdays',
               u'previous', u'risk', u'pdays_missing', u'job_admin.',
               u'job_blue-collar', u'job_entrepreneur', u'job_housemaid',
               u'job_management', u'job_retired', u'job_self-employed',
               u'job_services', u'job_student', u'job_technician', u'job_unemployed',
               u'job_unknown', u'marital_divorced', u'marital_married',
               u'marital_single', u'education_primary', u'education_secondary',
               u'education_tertiary', u'education_unknown', u'housing_no',
               u'housing_yes', u'loan_no', u'loan_yes', u'contact_cellular',
               u'contact_telephone', u'contact_unknown', u'month_apr', u'month_aug',
               u'month_dec', u'month_feb', u'month_jan', u'month_jul', u'month_jun',
               u'month_mar', u'month_may', u'month_nov', u'month_oct', u'month_sep',
               u'poutcome_failure', u'poutcome_other', u'poutcome_success',
               u'poutcome_unknown']
```

```python
from sklearn.cross_validation import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

kf = KFold(bank_data.shape[0], n_folds=10, shuffle=True)

k_list = []
knn_accuracy_list = []
knn_accuracy_std_list = []

for i in range(1,21):
    cv_accuracy_list = []
    for train_index, test_index in kf:
        cv_train, cv_test = bank_data.ix[train_index], bank_data.ix[test_index]
        knn_model = KNeighborsClassifier(n_neighbors = i).fit(cv_train[column_list],cv_train['resp'])
        bank_data['knn'+str(i)] = knn_model.predict_proba(bank_data[column_list])
        predicted_class_name = knn_model.predict(cv_test[column_list])
        cv_accuracy_list.append(accuracy_score(cv_test['resp'], predicted_class_name))
        
    k_list.append(i)
    knn_accuracy_list.append(np.mean(cv_accuracy_list))
    knn_accuracy_std_list.append(np.std(cv_accuracy_list))
    print i, np.mean(cv_accuracy_list), np.std(cv_accuracy_list)
```

```python
from sknn.mlp import Classifier, Layer

nn = Classifier(
    layers=[
        Layer("Rectifier", units=100),
        Layer("Softmax")],
    learning_rate=0.002,
    n_iter=1)

nn_model = nn.fit(bank_data[column_list], bank_data['resp'])

y_valid = nn_model.predict_proba(bank_data[column_list])
predicted_class_name = nn_model.predict(bank_data[column_list])

accuracy_score(bank_data['resp'], predicted_class_name)
```

```python
from sklearn.metrics import precision_recall_fscore_support

print precision_recall_fscore_support(bank_data['resp'], predicted_class_name, average='binary', pos_label=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_leaf=10, n_jobs=-1, 
                                   random_state=1234, verbose=0).fit(bank_data[column_list],bank_data['resp'])
rfc_score = rfc_model.predict_proba(bank_data[column_list])
predicted_class_name = rfc_model.predict(bank_data[column_list])
print accuracy_score(bank_data['resp'], predicted_class_name)
print precision_recall_fscore_support(bank_data['resp'], predicted_class_name, average='binary', pos_label=1)
print precision_recall_fscore_support(bank_data['resp'], predicted_class_name, average='binary', pos_label=0)
```
